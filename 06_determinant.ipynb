# Determinant

Every transformation squishes or stretches the space by some factor. The Determinant measures exactly this: how much a matrix (hence, a transformation) increases or decreases the area of a given region. Since every region in the space can be approximated by small enough squares, we can take the square having $\hat{i}$ and $\hat{j}$ as sides, without loss of generality. 

For example, let's take this matrix:

$$
\begin{pmatrix} 3 & 0 \\ 0 & 2 \end{pmatrix}
$$

This matrix scales $\hat{i}$ by a factor of $3$ and $\hat{j}$ by a factor of $2$. Therefore, tha square having $\hat{i}$ and $\hat{j}$ as sides, will become a square with an area of $6$. If we instead cosider the shear:

$$
\begin{pmatrix} 1 & 1 \\ 0 & 1 \end{pmatrix}
$$

This transformation leaves unchanged the area of the square having $\hat{i}$ and $\hat{j}$ as sides.

```python
import numpy as np
import utils.chart as utils

scale = np.array([
    [3, 0],
    [0, 2]
])

shear = np.array([
    [1, 1],
    [0, 1]
])

utils.visualize_transformation(scale, "This matrix will scale the size of the area by a factor of 6")
utils.visualize_transformation(shear, "This matrix will leave the area unchanged")
```

We call this scale factor the *determinant* of the transformation. So, if the transformation scales all areas by a factor of $3$, its determinant will be $3$. If a transformation squishes all areas by a factor of $2$, its determinant will be $1/2$. If the transformation squishes all the space onto a line or onto a single point, the determinant will be $0$. This means that some set of vectors is linear dependant if the matrix having these vector as columns has a null determinant. The determinant of a transformation is negative when we are "changing the side" from where we are seeing the Euclidian space. We use to think $\hat{i}$ on the right of $\hat{j}$. If a transformation is such that $\hat{i}$ will flip on the left of $\hat{j}$, then that transformation has a negative determinant. For the 3D case, we can point our fingers of the right hand in the 3 directions: the thumb on $\hat{k}$, the index finger on $\hat{i}$ and the middle finger on $\hat{j}$. If a transformation is such that, to do the same thing, we are coerced to use the left hand instead, than that transformation has a negative determinant.

## Compute a determinant

Given a $2 * 2$ matrix:

$$
M = \begin{pmatrix} a & b \\ c & d \end{pmatrix}
$$

the determinat will be:

$$
det(M) = ad - bc
$$

In fact, suppose that $b$ and $c$ are $0$, then the determinant will be $a * d$, which make sense because we are scaling $\hat{i}$ by a factor of $a$ and $\hat{j}$ by a factor of $d$. If we have $b$ **or** $c$ equals to $0$, that this make sense again, because we are applying a shear to the Euclidian space, which makes sqare areas become parallelepipeds areas of with the same area value. If both $b$ **and** $c$ differs from $0$, then we are stretching or squishing the square in the diagonal direction. When $bc$ becomes equal to $ad$, the determinant is $0$ and the Euclidian space is squished onto a line.

**theo:** $det(M_1 M_2) = det(M_1) * det(M_2)$

Justify the previous equation in terms of scaling and squishing areas is pretty trivial. If we scale and area by a factor of $2$, and then by a factor of $3$, we will have a total scale factor equal to $6$.
