#  Linear transformation interpretation

We said that:

$$
\vec{v} \times \vec{w} = \det(\begin{pmatrix} \hat{i} & v1 & w1 \\ \hat{j} & v2 & w2 \\ \hat{k} & v3 & w3 \end{pmatrix}) = \hat{i}(v2*w3 - w2*v3) + \hat{j}(-v1*w3 + w1*v3) + \hat{k}(v1*w2 - w1*v2)
$$

And we have three different numbers to represent a point in the standard basis. Remember that every trasformation $\mathbb{R}^n \rightarrow \mathbb{R}$ is represented by a $1*n$ matrix which, if transposed, is a vector (called *dual vector*) such that applying the linear transformation is the same as taking the dot product with that vector.

Let's:

1. Define a linear transformation $\mathbb{R}^3 \rightarrow \mathbb{R}$ in terms of two vector $\vec{v}$ $\vec{w}$.
2. Find its dual vector.
3. Show that this dual vector is $\vec{v} \times \vec{w}$.

### Step one

Let's define our linear transformation as a function $f: \mathbb{R}^3 \rightarrow \mathbb{R}$ that takes every point $\begin{pmatrix} x \\ y \\ z \end{pmatrix} \in \mathbb{R}^3$ and returns:

$$
f(\begin{pmatrix} x \\ y \\ z \end{pmatrix}) = \det(\begin{pmatrix} x & v_1 & w_1 \\ y & v_2 & w_2 \\ z & v_3 & w_3 \end{pmatrix})
$$

Can be demonstrated that this transformation is linear. Since is linear, we can find its dual vector.

### Step two

Which is the vector such that:

$$
\begin{pmatrix} ? & ? & ? \end{pmatrix} \begin{pmatrix} x \\ y \\ z \end{pmatrix} = \det(\begin{pmatrix} x & v_1 & w_1 \\ y & v_2 & w_2 \\ z & v_3 & w_3 \end{pmatrix})
$$

For the duality property, we can interpret the entire transformation as the dot product between the vectors:

$$
\begin{pmatrix} ? \\ ? \\ ? \end{pmatrix} \cdot \begin{pmatrix} x \\ y \\ z \end{pmatrix} = \vec{p} \cdot \vec{x} = \det(\begin{pmatrix} x & v_1 & w_1 \\ y & v_2 & w_2 \\ z & v_3 & w_3 \end{pmatrix})
$$

### Step three

So we are looking for the vector $\vec{p}$ such that taking the dot product between $\vec{p}$ and any other vector $\vec{x}$ gives the same result of that "determinant" where the first column are versors. Which gives us:

$$
p_1 = v2*w3 - w2*v3 \\
p_2 = -v1*w3 + w1*v3 \\
p_3 = v1*w2 - w1*v2
$$

Which is exactly the expression of the cross product.

Repeating: what vector $\vec{p}$ has the special property that when we take the dot product between $\vec{p}$ and any other vector in $\mathbb{R}^3$, it gives the same result as computing that strange determinant? Is the vector identified by the cross product.

## Geometric interpretation

Remember what we said about the dot product on the Euclidian space: it computes the area of the parallelogram having the two vector as sides. We computed it by putting two vector one on the side of the other and calculating the determinant. But what about $\mathbb{R}^3$? Well, if we were computing the area in the Euclidian space, we can think of computing the determinant of our 3 vector in a 3D space, as computing the area between these vectors.

So: What 3D vector $\vec{p}$ has the special property that, when we take a dot product between $\vec{p}$ and some other vector $\vec{x}$, it gives the same result as if we take the signed volume of a parallelepiped defined by this vector $\vec{x}$ along with $\vec{v}$ and $\vec{w}$?

Remeber that taking the dot product between two vector is the same as multiplying the lenght of the projection of the first on the second by the lenght of the second.

Now we want to compute the area of the parallelepiped defined by $\vec{v}$, $\vec{w}$ and $\vec{x}$. So we take the area of the parallelogram identified by $\vec{v}$ and $\vec{w}$ and multiply it not by $\vec{x}$ but by its component on the direction perpendicular to that parallelogram (this is how we get the volume). 

So the we can define our linear transformation as the transformation that takes the component of a given vector in the direction perpendicular to the parallelogram, and multiply it by the area of the parallelogram. But this is the same thing as taking a dot product between $\vec{x}$ and the vector perpedicular to the parallelogram.

So in this equation:

$$
\begin{pmatrix} p_1 \\ p_2 \\ p_3 \end{pmatrix} \cdot \begin{pmatrix} x \\ y \\ z \end{pmatrix} = \det(\begin{pmatrix} x & v_1 & w_1 \\ y & v_2 & w_2 \\ z & v_3 & w_3 \end{pmatrix})
$$

On the right side we project $\vec{x}$ onto some vector and we multiply it by the area of the parallelogram. On the left side we project $\vec{x}$ onto some vector and then we multiply the lenght of the two vectors. This means that this unknown vector has to be the vector that is perpendicular to the parallelogram (otherwise we wouldn't have the area).
