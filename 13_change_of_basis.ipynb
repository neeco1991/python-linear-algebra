# Change of basis

What happens if someone else is using a different coordinate system? You'd look at the same point on the Euclidian space but you'd give it different coordinates. For instance, suppose you are looking at a vector in the Euclidian space and you are giving it the coordinates: $\vec{p} = \begin{pmatrix} 3 \\ 2 \end{pmatrix}$, but someone else, looking at the same point, can give the coordinates $\vec{p'} = \begin{pmatrix} 5/3 \\ 1/3 \end{pmatrix}$. 

Suppose that you, looking at someone else basis vectors, discover that 

$$
\hat{i}' = \begin{pmatrix} 2 \\ 1 \end{pmatrix} \\
\hat{j}' = \begin{pmatrix} -1 \\ 1 \end{pmatrix}
$$

and he is trying to talk about a vector which in his coordinate system is  $\vec{p}' = \begin{pmatrix} -1 \\ 2 \end{pmatrix}$, how do you express that vector in your coordinate system? 

$$
\vec{p} = \begin{pmatrix} \hat{i}' & \hat{j}' \end{pmatrix} \vec{p}' = \begin{pmatrix} 2 & -1 \\ 1 & 1 \end{pmatrix} \begin{pmatrix} -1 \\ 2 \end{pmatrix} = \begin{pmatrix} -4 \\ 1 \end{pmatrix}
$$

So the matrix:

$$
A = \begin{pmatrix} 2 & -1 \\ 1 & 1 \end{pmatrix}
$$

is a way to move from our system with our basis vectors, to a new system with a new set of basis vectors. Or, in other words, is a way to translate what another person is saying to our language.

How to do the opposite? How do we translate what we are saying in the other person language? We need to build a matrix from their basis vectors to ours. That matrix is simply:

$$
A^{-1} = \begin{pmatrix} 1/3 & 1/3 \\ -1/3 & 2/3 \end{pmatrix}
$$

Hence, we can explain why when we looked at  $\vec{p} = \begin{pmatrix} 3 \\ 2 \end{pmatrix}$, the other person said that the coordinates were $\vec{p'} = \begin{pmatrix} 5/3 \\ 1/3 \end{pmatrix}$:

$$
A^{-1} \vec{p} = \begin{pmatrix} 1/3 & 1/3 \\ -1/3 & 2/3 \end{pmatrix} \begin{pmatrix} 3 \\ 2 \end{pmatrix} = \begin{pmatrix} 5/3 \\ 1/3 \end{pmatrix} = \vec{p'}
$$

## Transformations and change of basis

What if we want to talk about a transformation with a person with different coordinate system? Suppose we have a 90 degree counterclockwise rotation matrix like this:

$$
M = \begin{pmatrix} 0 & -1 \\ 1 & 0 \end{pmatrix}
$$

How would the other person describe the same transformation? 

First, we take a random point in the Euclidian space, written in the other person language:

$$
\vec{p}' = \begin{pmatrix} -1 \\ 2 \end{pmatrix}
$$

and translate it into our language using the change of basis matrix:

$$
\vec{p} = A \vec{p}' = \begin{pmatrix} 2 & -1 \\ 1 & 1 \end{pmatrix} \begin{pmatrix} -1 \\ 2 \end{pmatrix}
$$

Then, we apply the transformation matrix:

$$
\vec{p_2} = MA\vec{p} = \begin{pmatrix} 0 & -1 \\ 1 & 0 \end{pmatrix} \begin{pmatrix} 2 & -1 \\ 1 & 1 \end{pmatrix} \begin{pmatrix} -1 \\ 2 \end{pmatrix}
$$

This tells where the vector lands, but in our language. To translate to their language  we apply $A^{-1}$:

$$
\vec{p_2}' = A^{-1}MA\vec{p} = \begin{pmatrix} 1/3 & 1/3 \\ -1/3 & 2/3 \end{pmatrix} \begin{pmatrix} 0 & -1 \\ 1 & 0 \end{pmatrix} \begin{pmatrix} 2 & -1 \\ 1 & 1 \end{pmatrix} \begin{pmatrix} -1 \\ 2 \end{pmatrix}
$$

So, $A^{-1}MA$ is the transformation matrix in their language. $A$ and $A^{-1}$ are called empaty matrix.
