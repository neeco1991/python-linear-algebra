# Linear Combinations

Let's start with simple, human-readable terms. A *linear combination* of two vectors, is the sum of the two vectors, each one scaled by some factor. 


```python
# We generate the coordinate system:

import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.grid(True)
x_ticks = np.arange(-5, 6, 1)
y_ticks = np.arange(-5, 6, 1)
ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)
ax.set_xlabel("x", ha='right', x=1.0)
ax.set_ylabel("y", ha='right', y=1.0)

# And two vectors

v = np.array([1, 2])
w = np.array([-2, 0.5])
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label="v")
ax.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='y', label="w")

# Now we generate a couple of linear combinations l1 and l2

l1 = 0.5 * v + 1.5 * w # Here we scaled l1 by 0.5 and l2 by 1.5
l2 = 1.1 * v + 0.7 * w # Here we scaled l1 by 1.1 and l2 by 0.7
l3 = -1.3 * v + -1.7 * w # Here we scaled l1 by -1.3 and l2 by -1.7

ax.quiver(0, 0, l1[0], l1[1], angles='xy', scale_units='xy', scale=1, color='r', label="l1 = 0.5v + 1.5w", width=0.003)
ax.quiver(0, 0, l2[0], l2[1], angles='xy', scale_units='xy', scale=1, color='r', label="l2 = 1.1v + 0.7w", width=0.003)
ax.quiver(0, 0, l3[0], l3[1], angles='xy', scale_units='xy', scale=1, color='r', label="l3 = -1.3v + -1.7w", width=0.003)

ax.set_title("Linear combinations")
ax.legend()
```

What happens if we try to vary the scale factor of the two vectors? 

The set of all possible vectors, obtained by varing the two scale factors, is called *span* of the two vectors $\vec{v}$ and $\vec{w}$. In other words, the *span* generated by the two vectors is the set of all their possible linear combinations.

Now we can try to give a more technical definition of these two quantities:

**def:** given two vectors $\vec{v_1}$ and $\vec{v_2}$ and two scalars $a_1$ and $a_2$, we define a linear combination of $\vec{v_1}$ and $\vec{v_2}$ the value $a_1\vec{v_1} + a_2\vec{v_2}$.

**def:** given two vectors $\vec{v_1}$ and $\vec{v_2}$, we define *span of S* $S={\vec{v_1}, \vec{v_2}}$ the set of all possible linear combinations of $\vec{v_1}$ and $\vec{v_2}$: $span(\vec{v_1}, \vec{v_2}) := \{a_1\vec{v_1} + a_2\vec{v_2}\}$.

## Linear dependancy

In our previous example, the span of the two vectors coincides with the Euclidian space. In other words, every point in the Euclidian space, can be expressed as a linear combination of our two vectors $\vec{v}$ and $\vec{w}$. But this is not always true.

For instance, if we take the two vectors:

$$
\vec{v} = \begin{pmatrix} -1 \\ 1 \end{pmatrix},
\vec{w} = \begin{pmatrix} 2 \\ -2 \end{pmatrix}
$$

And we draw them in a chart, we notice that they lie on the same line. And so do their linear combinations!!!

```python
# Clean the chart
for collection in ax.collections[:]:
    collection.remove()


v = np.array([-1, 1])
w = np.array([2, -2])
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label="v")
ax.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='y', label="w")

l1 = 0.5 * v + 1.5 * w
l2 = 1.1 * v + 0.7 * w
l3 = -1.3 * v + -1.7 * w

ax.quiver(0, 0, l1[0], l1[1], angles='xy', scale_units='xy', scale=1, color='r', label="l1 = 0.5v + 1.5w", width=0.003)
ax.quiver(0, 0, l2[0], l2[1], angles='xy', scale_units='xy', scale=1, color='r', label="l2 = 1.1v + 0.7w", width=0.003)
ax.quiver(0, 0, l3[0], l3[1], angles='xy', scale_units='xy', scale=1, color='r', label="l3 = -1.3v + -1.7w", width=0.003)

ax.set_title("Linear combinations")
ax.legend()

# Display the chart again
fig
    
```

This happens because one of the two vectors can be expressed as a linear combination of the other. In fact:

$$
\vec{v_2} = -2\vec{v_1}
$$

Or, equivalently:


$$
\vec{v_1} = -0.5\vec{v_2}
$$


**def:** a set of vectors is said to be *linearly independent* if the only linear combination that equals the zero vector $\vec{0} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$ is the one with all the scales equal to $0$.

In our case $1\vec{v_1} + 0.5\vec{v_2} = 0$, so such linear combination exists and has non-null constants ($1$ and $0.5$). Therefore, our set of vector is *linear dependant*.

When a set of vector is linearly independent, they generate all the Euclidian space.

## Basis

We can finally give the definition of *basis of a vector space*:

**def:** a *basis* of a vector space $V\in{K^n}$ is a set of $n$ linear indipendent vectors.

Every element in $K^n$ can be expressed as a linear combination of the vectors in the basis.

**def:** the *standard basis* of a space vector, is the set of vector whose components are all $0$, except one that equals $1$.

In the Euclidian space, we use the standard basis: $\hat{i} = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$ and $\hat{j} = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$.

Given what we said about basis, any point $\vec{P}$ in the Euclidian space can be expressed as a linear combination of the standard basis: $\vec{P} = x\hat{i} + y\hat{j}$
