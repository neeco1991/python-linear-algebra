# Vectors

Vector are points in the n-dimensional space. We can think of them as an arrow starting from the origin of our coordinate system to a point in the n-space.

First of all, we create the coordinate system chart

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots() # Get both the figure and axes object

# --- Center the Axes (Spines) ---
# Hide the top and right spines
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

# Move the bottom and left spines to the center (0,0)
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 0))

# Set limits and add a grid
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.grid(True)

# --- Set Ticks for Each Unit ---
# Create arrays of numbers from -5 to 5 for the ticks
x_ticks = np.arange(-5, 6, 1)
y_ticks = np.arange(-5, 6, 1)

# Use ax.set_xticks and ax.set_yticks
ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)

# --- Add Title and Labels ---
# Use ax.set_title, ax.set_xlabel, etc.
ax.set_xlabel("x", ha='right', x=1.0)
ax.set_ylabel("y", ha='right', y=1.0)
```

Now we add a vector

```python

# --- Plotting ---
# Initialize a vector v
v = np.array([1, 2])

# Use ax.quiver to plot on your specific axes
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label="v")
ax.set_title("Vector [1, 2]")
ax.legend()

# Display the final, modified figure object
fig
```

## Sum of two vectors

The sum of two vectors `v = [v1, v2]` and `w = [w1, w2]` is equal to `v + w = [v1 + w1, v2 + w2]`.

We start adding w

```python
# --- Plotting ---
# Initialize a vector v
w = np.array([3, -1])

# Use ax.quiver to plot on your specific axes
ax.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='r', label="w")

ax.set_title("Vectors v and w")
ax.legend()

# Display the final, modified figure object
fig
```

Now we compute the sum:

```python
z = v + w

# Use ax.quiver to plot on your specific axes
ax.quiver(0, 0, z[0], z[1], angles='xy', scale_units='xy', scale=1, color='y', label="v + w")

ax.set_title("Sum of two vectors")
ax.legend()

# Display the final, modified figure object
fig
```
