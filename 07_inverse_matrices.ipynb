# Inverse Matrices


**def:** we define *linear equation* an equation where each term is either a constant or a constant multiplied by a single variable to the power of 1.

Example of linear equations are:

$$
x = 1 \\
x -2z = 7 \\
x + y + z = 0
$$

Example of non-linear equations are:

$$
x^2 = 2 \\
x + sin(x) = \pi \\
xy = 2
$$

**def:** we define *linear system of equations* a system of linear equations.

$$
\begin{cases}
2x + 3y = 7 \\
x - y = 1
\end{cases}
$$

We can package the previous system of linear equations in a *vector equation*:


$$
A\vec{x} = \vec{v} \Rightarrow \begin{pmatrix} 2 & 3 \\ 1 & -1 \end{pmatrix} * \begin{pmatrix} x \\ y \end{pmatrix} = \begin{pmatrix} 7 \\ 1 \end{pmatrix} 
$$

Solving this equation, is like finding the vector $\vec{x}$ that, after applying the transformation $A$, lands on $\vec{v}$.

There are two cases, one when $det(A) = 0$ and the transformation squishes all the point of the Euclidian space onto a line, and the case when $det(A) \neq 0$.

In the case when $det(A) \neq 0$, there will always be one and only one $\vec{x}$ that, after applying the transformation $A$, lands on $\vec{v}$, and we can find it solving the following:

$$
\vec{x} = A^{-1} * \vec{v}
$$

Where $A^{-1}$ is the *inverse matrix* of $A$. 

```python
import numpy as np

A = np.array([[2, 3], [1, -1]])
v = np.array([7, 1])

A_inverse = np.linalg.inv(A)

solution = np.matmul(A_inverse, v)

print(f"The solution of the equation is: {solution}")
```

Using the tranformation parallel, if $A$ is the transformation that rotates the Euclidian space by 90 degrees couterclockwise, then $A^{-1}$ will be the matrix that rotates the Euclidian space by 90 degrees clockwise. In general:

$$
A * A^{-1} = I = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}
$$

## Rank of a matrix

Everytime that the determinant of a matrix exists, the inverse of the matrix exists. However, when the determinant is equal to $0$ and the transformation squishes the solution space in a space of smaller dimension, the inverse of the transformation does not exists. For instance, you cannot unsquish a plane to turn into a volume. At least, that is not something that a *fuction* can do.

However, even if $A^{-1}$ doesn't exists, is still possible that a solution of $A\vec{x} = \vec{v}$ exists. In particular, that solutions exists if and only if the vector $\vec{v}$ lies on the line where all the Euclidian space was squished onto after applying the transformation.

If we think at the 3D case, a transformation $A$ with $det(A) = 0$ can squish the volume onto a plane, onto a line or onto a single point. When the output of a transformation is a line, we say that the transformation has *rank 1*, when is a plane the rank is 2, and so on. This obiously implies that a $2 * 2$ matrix has a rank not bigger that 2.

**def:** we call *rank* of a transformation the number of dimensions in the output of that transformation. 

**def:** the set of all possible outputs of $A\vec{v}$ is called the *column space* of $A$. In other words, the column space is the span of the column of the transformation matrix.

So we can give a better definition of *rank*:

**def:** we call *rank* of a transformation the number of dimensions in the column space.

When the rank is as high as it can be (when it equals the number of columns) we call it *full rank*.

The origin $\vec{0} = \begin{pmatrix} 0 \\ 0 \end{pmatrix}$ will always be included in the column space, since for a transformation to be linear must mantain the origin fixed. If the only vector such that $A\vec{x} = \vec{0}$ is the origin $\vec{0}$, the matrix has full rank. This is not true for matrices that has not full rank. If the rank of the transformation is one less the full rank, then there is a full line of vectors $\vec{x}$ such that $A\vec{x} = \vec{0}$, which means that all the vector on that line will be squished to the origin. If the rank of the transformation is two less than the full rank, then we will have a full plan of solution for the previous equation, so a full plan of vectors squished onto the origin.

**def:** The set of vector such that $A\vec{v} = \vec{0}$ is called the *null space* or the *kernel* of the transformation.

In the system of linear equation, the *kernel* gives us all the solution such that $A\vec{x} = \vec{0}$.
